<h5 style="font-family:Georgia;font-size:1.3vw; color:#000080">O que são opções? (resumidamente)</h5>

<div style="font-family:Georgia; font-size:1vw;color:#0000ff;">Opções são derivativos (derivam de um ativo) negociados em bolsa de valores, elas representam um contrato que dá ao seu titular o direito de comprar ou de vender um determinado ativo por um valor determinado em uma data específica do futuro. </div>

<strong style="font-family:Georgia; font-size:0.90vw;color:#000080;">Ativo-Objeto: Ativo do qual a opção se deriva</strong>

<strong style="font-family:Georgia; font-size:0.90vw;color:#000080;">Titular: Quem compra a opção</strong>

<strong style="font-family:Georgia; font-size:0.90vw;color:#000080;">Lançador: Quem vende a opção para o Titular</strong>

<strong style="font-family:Georgia; font-size:0.90vw;color:#000080;">Prêmio: valor pago por quem está comprando a opção</strong>

<strong style="font-family:Georgia; font-size:0.90vw;color:#000080;">Strike: Valor por qual a opção será exercida</strong>

<strong style="font-family:Georgia; font-size:0.90vw;color:#000080;">Vencimento: Data que ocorrerá o exercício das opções</strong>

<div style="font-family:Georgia; font-size:1vw;color:#0000ff;">Opções podem ser usadas para alavancagem, proteção e otimização de portfólio. Seja do lado do titular ou do lançador. </div>

<strong style="font-family:Georgia; font-size:0.90vw;color:#000080;">Existem opções para diversos mercados: Câmbio, Futuro, Ações e até de Commodities. E existem dois tipos de opções: Calls (opções de compra) e Puts (opções de venda).  </strong>

<div style="font-family:Georgia; font-size:1vw;color:#0000ff;">Uma analogia muito utilizada para entender as Puts são os seguros, o valor desembolsado permite ao segurado a garantia do valor integral (sinistro) do bem durante determinado período.</div>

<div style="font-family:Georgia; font-size:1vw;color:#0000ff;">Outra analogia utilizada para entender as Calls são os sinais dados na compra de um imóvel, que garantem a compra daquele bem até uma determinada data a um preço pré-determinado.</div>

<strong style="font-family:Georgia; font-size:0.90vw;color:#000080;">As opções possuem nomenclaturas diferentes para cada mês de vencimento e para cada tipo de opção. O vencimento ocorre sempre na 3ª Sexta-Feira do Mês.</strong>

<h5 style="font-family:Georgia;font-size:1.3vw; color:#006400">Como funciona o Covered Call ?</h5>

<div style="font-family:Georgia; font-size:1vw;color:#008000;">Essa operação estruturada se baseia na venda de uma call de um ativo e na manutenção de uma posição comprada nesse mesmo ativo. O nome call "coberta" vem exatamente do fato de se vender um dterminado ativo com a cobertura de outro ativo, que necessariamente anula a necessidade de chamada de margem para quem a executa.</div>

<h5 style="font-family:Georgia;font-size:1.3vw; color:#DAA520">Como é precificado a opção ?</h5>

<div style="font-family:Georgia; font-size:1vw;color:#DAA520;">Uma opção é precificada por meio do modelo de Black and Scholes. De acordo com as seguintes variáveis: Strike da opção, Preço Atual do ativo-objeto, taxa de juros livre de risco, tempo até o vencimento e a volatilidade implícita. </div>

<strong style="font-family:Georgia;font-size:1.00vw;">Importação das bibliotecas Scipy e Numpy para fazer a função do Black and Scholes</strong>

```python
import numpy as np
from scipy.stats import norm

N = norm.cdf
```

```python
def call_bs(S, K, T, r, sigma):
    d1 = (np.log(S/K) + (r + sigma**2/2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    return S * N(d1) - K * np.exp(-r*T)* N(d2)
```
